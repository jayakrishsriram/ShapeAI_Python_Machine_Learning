```python
import numpy as np
import sklearn 
import pandas as pd
from sklearn.datasets import load_boston
```

```python
df=load_boston()
df.keys()
```

```python
bosten=pd.DataFrame(df.data,columns=df.feature_names)
bosten.head()
```

```python
bosten['MEDV']=df.target
bosten.head()
```

```python
bosten.isnull()
```

```python
bosten.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x=bosten.drop('MEDV',axis=1)
y=bosten['MEDV']
X_train, X_test, Y_train, Y_test = train_test_split(x , y , test_size=0.15 , random_state=5)
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

lin_mod = LinearRegression()

lin_mod.fit(X_train, Y_train)
```

```python
y_train_predict=lin_mod.predict(X_train)
rmse=(np.sqrt(mean_squared_error(Y_train,y_train_predict)))
print("The model performance for training set")
print("RMSE is {}\n".format(rmse))
y_test_predict=lin_mod.predict(X_test)
rmse=(np.sqrt(mean_squared_error(Y_test,y_test_predict)))
print("The model performance for training set")
print("RMSE is {}\n".format(rmse))
```
